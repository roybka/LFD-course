```python
# Q 8,9,10
import numpy as np
import random as rn
import matplotlib.pyplot as plt

def sign(x):
    if x>0:
        return 1
    elif x==0:
        return 0
    elif x<0:
        return -1


def targetf(point):
    y=(point[1]**2+point[2]**2)-.6
    
    if sign(y)>0:
        return 1
    else: 
        return -1
def classify(w,point):
    outsum=0
    out=np.zeros(len(point))
    for i in range(len(point)):
        out[i]=w[i]*point[i]
        outsum+=out[i]
#         outsum=outsum+self.w0
    return outsum

npts=1000
result_ints=[]
tresult_ints=[]
ws=[]
for kk in range(1000):
    points=[[1,rn.random()*2-1,rn.random()*2-1] for x in range(npts) ]
    newpoints=[[x1,x2,x3,x2*x3,x2**2,x3**2] for x1,x2,x3 in points]
    outputs=np.array([targetf(k) for k in points])
    # inds=slice(rn.sample(range(npts),int(npts/10)))
    inds=np.index_exp[rn.sample(range(npts),int(npts/10))]
    outputs[inds]=outputs[inds]*-1

    # for i in range(len(points)):
    #     y=(outputs[i])
    #     if y>0:
    #         plt.scatter(points[i][1],points[i][2],marker='o')
    #     else:
    #         plt.scatter(points[i][1],points[i][2],marker='x')
    X=np.matrix(newpoints)
    Xpinv=np.linalg.pinv(X)
    Z=np.inner( Xpinv,outputs)
    q=np.array(Z[0])
    w=q[0]
    intcnt=0
    tintcnt=0
    for k in range(npts):
    #         for i in range(npts):
    #         i=rn.randint(0,npts-1)
        y1=classify(w,newpoints[k])
        y=sign(y1)

        if y==outputs[k]:
            pass
        else:
            intcnt+=1
    #             pt.w=np.add(  pt.w,[outputs[i]*points[i][0],outputs[i]*points[i][1]])
    #             pt.w0=pt.w0+outputs[i]
    tpoints=[[1,rn.random()*2-1,rn.random()*2-1] for x in range(npts) ]
    tnewpoints=[[x1,x2,x3,x2*x3,x2**2,x3**2] for x1,x2,x3 in tpoints]
    outputs=np.array([targetf(k) for k in tpoints])
    for k in range(npts):
#         for i in range(npts):
#         i=rn.randint(0,npts-1)
        y1=classify(w,tnewpoints[k])
        y=sign(y1)

        if y==outputs[k]:
            pass
        else:
            tintcnt+=1
    
    result_ints.append(intcnt)
    tresult_ints.append(tintcnt)
    ws.append(w)
in_err=[result_ints[x]/npts for x in range(len(result_ints)) ]
out_err=[tresult_ints[x]/npts for x in range(len(tresult_ints)) ]
np.average(in_err)
```
